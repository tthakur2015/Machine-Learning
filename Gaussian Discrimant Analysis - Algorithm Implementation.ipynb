```python
# Create GDA class

import numpy as np

class GDA:
    
    # Instance attribute
    def __init__(self, train_features, train_labels):
        self.train_features = train_features
        self.train_labels = train_labels
        
    # True and false label subsets
        true_subset = []
        false_subset = []
    
    # Calculating class prior probabilities
        true_count = 0
        false_count = 0
        
        for i,j in zip(self.train_features, self.train_labels):
            if j == 0:
                false_count += 1
                false_subset.append(i)
            else:
                true_count += 1
                true_subset.append(i)
        
        self.true_subset = np.array(true_subset)
        self.false_subset = np.array(false_subset)
        
        self.true_prior = true_count/len(train_features)
        self.false_prior = false_count/len(train_features)
        
    # Calculate feature means for each label class
        self.true_feature_means = np.mean(true_subset, axis = 0)
        self.false_feature_means = np.mean(false_subset, axis = 0)
    
    # Calculating covariance matrix of training features. Result should be a 30 x 30 matrix
        train_features_reshaped = np.reshape(train_features, (train_features.shape[1], train_features.shape[0]))
        self.cov_matrix = np.cov(train_features_reshaped, bias = True)
        
    def gaussian_pdf(self, x, mu, cov):
        mean_diff = x - mu
        
        
        
    
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import MinMaxScaler

cancer_data = load_breast_cancer()

examples = np.array(cancer_data.data)
labels = np.array(cancer_data.target)

scaler = MinMaxScaler()
examples = scaler.fit_transform(examples)
```

```python
model_gda = GDA(examples, labels)
```
